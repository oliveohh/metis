## Exploring the impact of extreme weather on MTA station ridership
By: Olivia Offutt

### ABSTRACT
Climate change is increasing extreme weather patterns in NYC, and the Extreme Weather Task Force was recently launched to improve citywide planning and response to these events. The Task Force needs transit data on storm system events to target MTA capital improvements and emergency planning strategies. This exploratory data analysis uses MTA turnstile data and National Weather Service storm warning data to identify ridership trends for the 2021 hurricane season (August 1 - October 31) and identify specific stations that saw the heaviest impacts to ridership both during storm events (acute station impact), and after storm events (indicating extensive storm damage). 

### DESIGN
Aggregating MTA turnstile data and National Weather Service by week, this analysis observes trends of ridership around storms for work weeks and weekends, as well as identifying the top ten stations that experienced significant ridership drops both acutely (during storm weeks) and chronically (enduring post storm weeks). This provides a starting point for identifying stations that should be prioritized for capital improvement planning to prevent future station service impacts as extreme weather becomes more common.

### DATA
- The NY MTA data represents cumulative turnstile entry data by station and date for peak storm season in New York (August 1 - October 31, 2021).  
- The National Weather Service Data represents official flash flood warnings for the NY counties that have MTA stations during the same storm season time period. 

### TOOLS
- SQLite and SQLAlchemy (data import)
- Pandas (data exploration / analysis)
- Matplotlib and Seaborn (data visualization)

### ALGORITHMS
- Description data for flood events by NY county to identify counties that might be more vulnderable to dangerous flash flood events.
- Aggregating turnstile data by station, week, and weekday vs weekend data.
- Identifying storm weeks and non-storm weeks for week by week total entries comparison.
- Developing line plots to show total MTA entries by weekend (Su-Sa) and work week (M-F) over the hurricane season and identify any trends during and after storm weeks.
- Developing weekend and work week weekly entry averages to compare non-storm and storm week entry averages.
- Drilling down on the three weeks before, during, and after storm Ida to develop percent changes in ridership comparing before- and during-storm weeks (showing acute impact), and before- and after-storm weeks (showing more long-term impact).
